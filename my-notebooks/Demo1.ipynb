```python
plt.figure(figsize = (18,7))

x_data = %sql SELECT distance FROM rides WHERE NOT distance = 0 LIMIT 10000
y_data = %sql SELECT duration FROM rides LIMIT 10000

plt.plot(x_data, y_data, 'og')

plt.ylabel('Distance')
plt.xlabel('Duration')
plt.title('Distance vs Duration')

x = []
for item in x_data:
    x.append(item[0])
    
for item in y_data:
    y.append(item[0])
    
x = np.array(x)
y = np.array(y)
m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x + b)
```
